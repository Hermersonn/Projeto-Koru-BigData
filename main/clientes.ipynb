```python
import pandas as pd

caminho_Df_1 = 'DataSet_brazilian-ecommerce/olist_customers_dataset.csv'

df_1 = pd.read_csv(caminho_Df_1)

display(df_1)
```

```python
# Renomeando colunas
df_1.rename(columns={'customer_id':'ID do Cliente','customer_unique_id':'ID Exclusivo do Cliente','customer_zip_code_prefix':'CEP','customer_city':'Cidade','customer_state':'Estado'}, inplace=True)

#Criando uma coluna para separar clientes por região
#criando um dicionário chamado regiao que mapeia os estados para as regiões correspondentes
regiao = {"AC": "Norte",
    "AL": "Nordeste",
    "AP": "Norte",
    "AM": "Norte",
    "BA": "Nordeste",
    "CE": "Nordeste",
    "DF": "Centro-Oeste",
    "ES": "Sudeste",
    "GO": "Centro-Oeste",
    "MA": "Nordeste",
    "MG": "Sudeste",
    "MT": "Centro-Oeste",
    "MS": "Centro-Oeste",
    "PA": "Norte",
    "PB": "Nordeste",
    "PE": "Nordeste",
    "PI": "Nordeste",
    "PR": "Sul",
    "RJ": "Sudeste",
    "RN": "Nordeste",
    "RO": "Norte",
    "RR": "Norte",
    "RS": "Sul",
    "SC": "Sul",
    "SE": "Nordeste",
    "SP": "Sudeste",
    "TO": "Norte",
}
#criando uma nova coluna chamada "Região" no DataFrame, preenchendo os valores com base no mapeamento do 
# estado para a região.
df_1["Região"] = df_1["Estado"].map(regiao)
```

```python
# A linha abaixo traz as informaçoes do dataset
df_1.info()

#Procurando valores nulos
print('\n')
print("Valores nulos")
valores_nulos = df_1.isnull().isna()
print(valores_nulos.any().any())
```
